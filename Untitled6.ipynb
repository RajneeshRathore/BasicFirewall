```python
import ast
```

```python
def extract_genres(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movies["genres"]= movies["genres"].apply(extract_genres)
```

```python
movies.shape
```

```python
movies.head(1)
```

```python
movies["keywords"]= movies["keywords"].apply(extract_genres)
```

```python
movies
```

```python
movies["cast"].values[0]
```

```python
def extract_cast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
      if count != 3:
        count+=1
        L.append(i["name"])
    return L
```

```python
movies["cast"] = movies["cast"].apply(extract_cast)
```

```python
movies
```

```python
movies["crew"].values[0]
```

```python
def extract_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"].lower() == "director":
            L.append(i["name"])
    return L
```

```python
movies["crew"]= movies["crew"].apply(extract_crew)
```

```python
movies
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies
```

```python
movies["genres"].head(1)
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies["overview"] = movies["overview"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

```python
movies["tags"] = movies["overview"]+movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies
```

```python
df = movies[["id","title","tags"]]
```

```python
df.head(1)
```

```python
df["tags"]=df["tags"].apply(lambda x:" ".join(x))
```

```python
df.head(1)
```

```python
df["tags"]=df["tags"].apply(lambda x:x.lower())
```

```python
df["tags"][0]
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# ps.stem("dancing")
def stem(txt):
  L = []
  for i in txt.split():
    L.append(ps.stem(i))
  return L
```

```python
df["tags"] = df["tags"].apply(stem)
```

```python
df.head(1)
```

```python
df["tags"] = df["tags"].apply(lambda x:" ".join(x))
```

```python
df.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
cv
```

```python
vector = cv.fit_transform(df["tags"]).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs = cosine_similarity(vector)
```

```python
cs[0]
```

```python
sorted(list(enumerate(cs[0])),reverse=True,key=lambda x:x[1] )[1:6]
```

```python
df[df["title"]=="Avatar"].index[0]
```

```python
df.iloc[1214].title
```

```python
def recommend(movie):
  L = []
  movie_idx = df[df["title"]==movie].index[0]
  movie_list = sorted(list(enumerate(cs[movie_idx])),reverse=True,key=lambda x:x[1] )[1:6]
  for i in movie_list:
    L.append(df.iloc[i[0]].title)
  return L
```

```python
print(recommend("Avatar"))
```
